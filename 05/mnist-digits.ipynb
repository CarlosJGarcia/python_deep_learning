```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
from tensorflow.keras.utils import to_categorical

mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

print("Muestro las 5 primeras im√°genes de entrenamiento:")
for n in range(0,5):
    print(f"Train image #{n} - label: {y_train[n]}")
    plt.imshow(x_train[n], cmap=plt.cm.binary)
    plt.show()

print("Muestro las 5 primeras im√°genes de test:")
for n in range(0,5):
    print(f"Test image #{n} - label: {y_test[n]}")
    plt.imshow(x_test[n], cmap=plt.cm.binary)
    plt.show()


# Normalizacion
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255

# Reshape de 60.000 28x28 a 60.000 784
x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)

y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)

print("Muestro las 5 primeras im√°genes de entrenamiento:")
for n in range(0,5):
    print(f"Train image #{n} - label: {y_train[n]}")

print("Muestro las 5 primeras im√°genes de test:")
for n in range(0,5):
    print(f"Test image #{n} - label: {y_test[n]}")

```
